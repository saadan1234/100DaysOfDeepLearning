# Library Imports

```python
from keras.models import Model
from keras.layers import *
from keras.utils import plot_model, model_to_dot
```

# Sample Neural Network

```python
x = Input(shape=(3,), name='Input')

# Sequential Layers
layer1 = Dense(128, activation='relu', name='Layer1')(x)
layer2 = Dense(64, activation='relu', name='Layer2')(layer1)

# Non-linearly connected layers
output1 = Dense(1, activation='linear', name='Regression')(layer2) # Regression Prediction
output2 = Dense(1, activation='sigmoid', name='Classification')(layer2) # Classification Prediction
```

```python
model = Model(inputs=x , outputs=[output1, output2])
model.summary()
```

# Model Visualization

```python
plot_model(model, show_shapes=True, show_layer_names=True, show_layer_activations=True, dpi=100)
```
