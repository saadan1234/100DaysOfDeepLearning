```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.models import Sequential
```

# Data Generation

```python
X,y = make_moons(n_samples=250, noise=0.05, random_state=42)
plt.scatter(X[:,0],X[:,1], c=y, s=100)
plt.show()
```

# Model Architecture

```python
model = Sequential()
model.add(Dense(10,activation='sigmoid',input_dim=2))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

# Batch Gradient Descent

```python
import time
from keras.models import clone_model

# Clone the original model
bgd_model = tf.keras.models.clone_model(model)
bgd_model.set_weights(model.get_weights())
bgd_model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

start = time.time()
history = bgd_model.fit(X,y,epochs=100,batch_size=200,validation_split=0.2)
print(time.time() - start)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
```

# Stochatic Gradient Descent

```python
# Clone the original model
sgd_model = tf.keras.models.clone_model(model)
sgd_model.set_weights(model.get_weights())
sgd_model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

start = time.time()
history = sgd_model.fit(X,y,epochs=100,batch_size=1,validation_split=0.2)
print(time.time() - start)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
```

# Mini batch Gradient Decsent

```python
# Clone the original model
mbgd_model = tf.keras.models.clone_model(model)
mbgd_model.set_weights(model.get_weights())
mbgd_model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

start = time.time()
history = mbgd_model.fit(X,y,epochs=100,batch_size=20,validation_split=0.2)
print(time.time() - start)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['loss'])
```

# Conclusion

1) Batch Gradient decsent in computationally faster, but slower in convergence and has smooth loss curve.

2) Stochastic Gradient decsent in computationally slower, but faster in convergence and has smooth loss curve.
