# Introduction

This notebook provides basic intuitionof how the classification algortihms learn to predict the classification margin and update its values.

# Data Creation

```python
from sklearn.datasets import make_classification
import numpy as np

x, y = make_classification(n_samples=200, n_features=2,n_informative=2,n_redundant=0, n_classes = 2, n_clusters_per_class=2, random_state = 42,hypercube=False,class_sep=15)
```

```python
import matplotlib.pyplot as plt
```

# Visualization

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=20)
```

# Activation Function

```python
def step(z):
    return 1 if z>0 else 0
```

# Learning Rule

```python
def classify(x, y):
  x = np.insert(x,2,1,axis=1)
  weights = np.ones(x.shape[1])
  alpha = 0.1

  for i in range(1000):
    point = np.random.randint(0,200)
    pred = step(np.dot(x[point],weights))
    weights = weights + alpha * (y[point] - pred) * x[point]
  return weights[2],weights[0:2]

```

```python
bias,weights = classify(x,y)
print(bias)
print(weights)
```

# Classification Line

```python
m = -(weights[0]/weights[1])
b = -(bias/weights[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

Visualization

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=20)
plt.ylim(-3,2)
```
